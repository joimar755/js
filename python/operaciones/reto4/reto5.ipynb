```python
import pandas as pd

df = pd.read_csv('Listado_Colegios.csv')

print('cantidad de filas y columnas del dataset inicial',df.shape)
df_limpio = df.drop_duplicates()
print('cantidad de filas y columnas del dataset limpio',df_limpio.shape)
print('cantidad de nulos por columnas')
print(df_limpio.isnull().sum())

print("Nombres de las columnas")
nombre_columnas=df_limpio.columns
print(nombre_columnas)

df_limpio=df_limpio.dropna(subset=['telefono'])
df_limpio['etnias']=df_limpio['etnias'].fillna('No Aplica')
df_limpio=df_limpio.drop(columns="sector")
df_limpio=df_limpio.drop(columns="genero")
df_limpio['niveles']=df_limpio['niveles'].fillna('No especificado')
df_limpio['jornadas']=df_limpio['jornadas'].fillna('No especificado')
df_limpio=df_limpio.drop(columns="caracter")
df_limpio['especialidad']=df_limpio['especialidad'].fillna('No especificado')
df_limpio['jornadas']=df_limpio['jornadas'].fillna('No especificado')
df_limpio['grados']=df_limpio['grados'].fillna('No especificado')
df_limpio['modelos_Educativos']=df_limpio['modelos_Educativos'].fillna('No especificado')
df_limpio['capacidades_Excepcionales']=df_limpio['capacidades_Excepcionales'].fillna('No especificado')
df_limpio['discapacidades']=df_limpio['discapacidades'].fillna('No especificado')
df_limpio=df_limpio.drop(columns="internado")
df_limpio=df_limpio.drop(columns="estado")
df_limpio=df_limpio.drop(columns="estrato_Socio_Economico")
df_limpio['correo_Electronico']=df_limpio['correo_Electronico'].fillna('No especificado')
print(f"Avance: {df_limpio.shape}")
print(df_limpio.isnull().sum())

df_limpio.to_csv('Listado_Colegios_Limpio.csv', index=False)

df_limpio.columns = df_limpio.columns.str.lower()
df_limpio.describe()
```
