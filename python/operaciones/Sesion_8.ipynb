Otros graficos


```python
import matplotlib.pyplot as plt
import numpy as np

np.random.seed(19680801)

def imagen_degradado(ax, direccion=0.3, rango_cmap=(0, 1), extent=None, **kwargs):
    """
    Dibuja una imagen con degradado basada en un mapa de colores.

    Parámetros
    ----------
    ax : Axes
        Los ejes en los que se dibujará.
    direccion : float
        La dirección del degradado. Este número está en el rango de 0 (=vertical) a 1 (=horizontal).
    rango_cmap : float, float
        La fracción (cmin, cmax) del mapa de colores que se utilizará para el degradado,
        donde el mapa de colores completo está en (0, 1).
    extent : tuple or None
        Los límites del área en la que dibujar la imagen.
    **kwargs
        Otros parámetros se pasan a `.Axes.imshow()`. En particular, *cmap*, *extent* y *transform* pueden ser útiles.
    """
    phi = direccion * np.pi / 2
    v = np.array([np.cos(phi), np.sin(phi)])
    X = np.array([[v @ [1, 0], v @ [1, 1]],
                  [v @ [0, 0], v @ [0, 1]]])
    a, b = rango_cmap
    X = a + (b - a) / X.max() * X
    im = ax.imshow(X, interpolation='bicubic', clim=(0, 1),
                   aspect='auto', extent=extent, **kwargs)
    return im

def barra_degradado(ax, x, y, ancho=0.5, fondo=0):
  for izquierda, arriba in zip(x, y):
    derecha = izquierda + ancho
    imagen_degradado(ax, extent=(izquierda, derecha, fondo, arriba),
                         cmap=plt.cm.Blues_r, rango_cmap=(0, 0.8))

fig, ax = plt.subplots()
ax.set(xlim=(0, 10), ylim=(0, 1))

imagen_degradado(ax, direccion=1, extent=(0, 1, 0, 1), transform=ax.transAxes,
                 cmap=plt.cm.RdYlGn, rango_cmap=(0.2, 0.8), alpha=0.5)

N = 10
x = np.arange(N) + 0.15
y = np.random.rand(N)
barra_degradado(ax, x, y, ancho=0.7)
plt.show()
```

Diagrama de tallo

```python
import matplotlib.pyplot as plt
import numpy as np

x = np.linspace(0.1, 2 * np.pi, 41)
y = np.exp(np.sin(x))

plt.stem(x, y)
plt.xlabel('x')
plt.ylabel('y')
plt.title('Gráfico de tallos')
plt.show()
```

HISTOGRAMAS

```python
import matplotlib.pyplot as plt
import numpy as np

np.random.seed(19680801)

n_bins = 10
x = np.random.randn(1000, 3)

fig, ((ax0, ax1), (ax2, ax3)) = plt.subplots(nrows=2, ncols=2)

colors = ['red', 'tan', 'lime']
ax0.hist(x, n_bins, density=True, histtype='bar', color=colors, label=colors)
ax0.legend(prop={'size': 10})
ax0.set_title('bars with legend')
ax1.hist(x, n_bins, density=True, histtype='bar', stacked=True)
ax1.set_title('stacked bar')
ax2.hist(x, n_bins, histtype='step', stacked=True, fill=False)
ax2.set_title('stack step (unfilled)')

x_multi = [np.random.randn(n) for n in [10000, 5000, 2000]]
ax3.hist(x_multi, n_bins, histtype='bar')
ax3.set_title('different sample sizes')

fig.tight_layout()
plt.show()
```

Gráfico de dispersión con diseño.

```python
import seaborn as sns
import matplotlib.pyplot as plt

sns.set(style="whitegrid")
datos = sns.load_dataset("tips")

sns.scatterplot(x="total_bill", y="tip", data=datos, color="blue", s=100)
plt.title('Relación entre la cuenta total y la propina', fontsize=14)
plt.xlabel('Cuenta total (USD)', fontsize=12)
plt.ylabel('Propina (USD)', fontsize=12)
plt.show()
```

Gráfico de dispersión con 3 variables y línea de tendencia

```python
import seaborn as sns
import matplotlib.pyplot as plt

datos = sns.load_dataset("tips")

sns.scatterplot(x='total_bill', y='tip', size='size', data=datos, hue='size', palette="cool", sizes=(20, 200))
sns.regplot(x='total_bill', y='tip', data=datos, scatter=False, color='red')

plt.title('Relación entre cuenta total, propina y tamaño del grupo')
plt.xlabel('Cuenta total (USD)')
plt.ylabel('Propina (USD)')
plt.show()
```

Mapa de calor

```python
import seaborn as sns
import matplotlib.pyplot as plt

sns.set(style="white")
datos = sns.load_dataset("flights")
tabla_pivote = datos.pivot(index="month", columns="year", values="passengers")
sns.heatmap(tabla_pivote, annot=True, fmt="d", cmap="YlOrRd", linewidths=0.5)
plt.title('Número de pasajeros por mes y año', fontsize=14)
plt.xlabel('Año', fontsize=12)
plt.ylabel('Mes', fontsize=12)
plt.show()
```
