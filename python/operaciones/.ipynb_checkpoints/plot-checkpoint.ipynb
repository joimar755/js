```python
import matplotlib.pyplot as plt
# Datos de ejemplo
meses = ['Enero', 'Febrero', 'Marzo', 'Abril', 'Mayo']
ventas = [1500, 1800, 1200, 1700, 2100]
# Crear el gráfico de líneas con color
plt.plot(meses, ventas, color='blue', marker='o', linestyle='--')
# Etiquetas en español
plt.xlabel('Meses')
plt.ylabel('Ventas en USD')
plt.title('Ventas Mensuales')
# Mostrar el gráfico
plt.show()
```

```python
import matplotlib.pyplot as plt
productos = ['Producto A', 'Producto B', 'Producto C']
cantidad_vendida = [50, 75, 100]
# Crear gráfico de barras con color
plt.bar(productos, cantidad_vendida, color='green')
# Etiquetas en español
plt.xlabel('Productos')
plt.ylabel('Cantidad Vendida')
plt.title('Ventas por Producto')
# Mostrar el gráfico
plt.show()
```

```python
import matplotlib.pyplot as plt
# Datos de ejemplo
horas_estudio = [1, 2, 3, 4, 5]
puntuaciones = [50, 60, 70, 80, 90]
# Crear gráfico de dispersión con color
plt.scatter(horas_estudio, puntuaciones, color='red')
# Etiquetas
plt.xlabel('Horas de Estudio')
plt.ylabel('Puntuación en el Examen')
plt.title('Relación entre Estudio y Resultados')
# Mostrar el gráfico
plt.show()
```
