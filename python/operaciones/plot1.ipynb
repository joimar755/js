```python
import matplotlib.pyplot as plt
prductos = ['A', 'B', 'C', 'D', 'E', 'F', 'G', 'H']
Bogota = [150,100,90,80,70,60,50,40]
Cali = [200,180,150,140,130,120,110,100]
Bucaramanga = [250,220,300,280,260,240,220,200]

plt.bar(prductos, Bogota, color='blue', label='Bogota',)
plt.bar(prductos, Cali, color='red', label='Cali', bottom=Bogota)
plt.bar(prductos, Bucaramanga, color='green', label='Bucaramanga', bottom=[i+j for i,j in zip(Bogota, Cali)])
plt.xlabel('Productos')
plt.ylabel('Ventas (unidades)')
plt.title('Ventas por ciudad')
plt.legend()
plt.show()
```

```python
import matplotlib.pyplot as plt
import seaborn as sns
import pandas as pd

datos = pd.DataFrame({
'Mes': ['Enero', 'Febrero', 'Marzo', 'Abril', 'Mayo', 'Junio',
'Julio', 'Agosto', 'Septiembre',
'Octubre', 'Noviembre', 'Diciembre'],
'SUV': [120, 130, 135, 140, 150, 160, 170, 180, 175, 185, 190, 200],
'Sedan': [100, 90, 95, 110, 120, 130, 140, 145, 150, 160, 170, 180],
'Hatchback': [80, 85, 90, 95, 100, 105, 110, 115, 120, 125, 130, 135]
})
sns.set(style="whitegrid")
plt.plot(datos['Mes'], datos['SUV'], color='blue', marker='o', label='SUV', linestyle='--')
plt.plot(datos['Mes'], datos['Sedan'], color='red', marker='o', label='Sedan', linestyle='--')
plt.plot(datos['Mes'], datos['Hatchback'], color='green', marker='o', label='Hatchback', linestyle='--')
plt.xlabel('Meses', fontsize=14)
plt.ylabel('Ventas (unidades)', fontsize=14)
plt.xticks(rotation=45, ha='right')
plt.legend()
plt.title('Evolución de ventas de automóviles por tipo', fontsize=16)

#print(datos)
```
