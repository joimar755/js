```python
import pandas as pd 
df = pd.read_csv('Listado_Colegios.csv')

print('primeras filas del dataset')
print(df.head())

print('*******************************')
print('informacion de la estructura del dataset')
print(df.info())
print('*******************************')
print('primeras 10 filas del dataset')
print(df.head(10))
print('*******************************')
print('Ultimas 10 filas del dataset')
print(df.tail(10))
print('*******************************')
print('nombres de las columnas')
print(df.columns)
print('*******************************')
print('valores unicos')
print(df.nunique())
print('*******************************')
print('mostrar algunas filas que se encuentra duplicadas')
filas_duplicadas = df[df.duplicated()]
if not filas_duplicadas.empty:
    print(filas_duplicadas.head())
else:
    print('No hay filas duplicadas en el dataset')
print('*******************************')
print('cantidad e filas con valores nulos')
print(df.isnull().sum())
print('*******************************')
print('porcentajes de valores nulos por columna')
print(df.isnull().sum() / len(df) * 100)
print('*******************************')
columnas_con_nulos = df.columns[df.isnull().any()]
print('Columnas con valores nulos:')
print(columnas_con_nulos)
print('*******************************')
print('datos unicos en el dataset')
print(df.dtypes.value_counts())
print('*******************************')
print('cantidad de columnas por tipo de texto')
columnas_tipos_objeto= df.select_dtypes(include=['object']).columns
print(columnas_tipos_objeto)
print('*******************************')
print('ejemplo de valores en columnas de tipo caracter')
for col in columnas_tipos_objeto:
    print('ejemplo de valores en la columna {col}')
    print(df[col].dropna().unique()[:5])
print('*******************************')
columnas_numericas = df.select_dtypes(include=['int64', 'float64']).columns
print('Columnas numericas:')

for col in columnas_numericas:
    print(f'rango de datos de columnas numericas')
    #print(f'Valores unicos: {df[col].nunique()}')
    print(f'Minimo: {df[col].min()}')
    print(f'Maximo: {df[col].max()}')
    print(f'Media: {df[col].mean()}')
    #print(f'Desviacion estandar: {df[col].std()}')
    
```
