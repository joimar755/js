```python
from matplotlib import colors
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

df = pd.read_csv('Listado_Colegios_Limpio.csv')
plt.figure(figsize=(6, 6))
distribucion =df['tipo_Establecimiento'].value_counts()
distribucion.plot(kind='bar', color=['#FF9999', '#66B3FF', '#99FF99', '#FFCC99'],edgecolor='black')
plt.title('Distribucin de Establecimientos',fontsize=16, weight='bold')
plt.xlabel('Tipo de Establecimiento',fontsize=14, weight='bold')
plt.ylabel('Cantidad de colegios',fontsize=12, weight='bold')
plt.tight_layout()
plt.show()

plt.figure(figsize=(6, 6))
zona = df['zona'].value_counts()
colores = ['#FF9999', '#66B3FF', '#99FF99', '#FFCC99']
explode = (0.1, 0.1, 0.1)  # Match the number of unique zonas
zona.plot(kind='pie', autopct='%1.1f%%', startangle=90, colors=colores[:len(zona)], explode=explode, shadow=True)
plt.title('Distribucin de Establecimientos por Zona urbana', fontsize=16, weight='bold')
plt.ylabel('')
plt.tight_layout()
plt.show()
```
