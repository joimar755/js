```python
%pip install matplotlib --quiet

import pandas as pd
import numpy as np
from matplotlib import pyplot as plt

df_financiero = pd.read_csv('datos_financieros.csv')
print('resumen de los datos financieros:')
print(df_financiero.describe())
print('datos nulos:')
print(df_financiero.isnull().sum())
df_financiero['Salario_Anual'].fillna(df_financiero['Salario_Anual'].median())
df_financiero['Deuda'].fillna(df_financiero['Deuda'].median())
df_financiero.to_csv('datos_financieros_limpios.csv', index=False)

plt.figure(figsize=(10, 6))
plt.scatter(df_financiero['Edad'],df_financiero['Deuda'], c='blue', alpha=0.5)
plt.title('Deuda vs Edad',fontsize=16)
plt.xlabel('Edad',fontsize=14)
plt.ylabel('Deuda',fontsize=14)
plt.grid()

x=np.polyfit(df_financiero['Edad'],df_financiero['Deuda'],1)
p=np.poly1d(x)
plt.plot(df_financiero['Edad'],p(df_financiero['Edad']),color='red', label='Tendencia')
plt.legend()
plt.show()

plt.figure(figsize=(9,6))
df_financiero['Historial_Crediticio'].value_counts().plot(kind='bar', color='orange', edgecolor='black')
plt.title("Distribucin de Historial Crediticio ", fontsize=15)
plt.xlabel("Historial crediticio", fontsize=12)
plt.ylabel("Cantidad de clientes ", fontsize=12)
plt.grid(axis='y',linestyle='--',alpha=0.7)
plt.show()
```
