```python
import pandas as pd
import numpy as np

# Leer el CSV
df = pd.read_csv('DatosPersonales.csv')
print(df)
# Reemplazar valores nulos representados de diferentes formas
df.replace([None, 'nan', '-'], np.nan, inplace=True)

# Correcciones sin inplace en columnas
df['Genero'] = df['Genero'].replace({'M': 'Masculino', 'H': 'Masculino', 'F': 'Femenino', 'Mujer': 'Femenino'})
df['Edad'] = df['Edad'].fillna(df['Edad'].median())
df['Ingresos'] = df['Ingresos'].fillna(df['Ingresos'].median())
df['Pais'] = df['Pais'].fillna(df['Pais'].mode()[0])
df['Visitas'] = df['Visitas'].astype(float)
df['Visitas'] = df['Visitas'].fillna(int(df['Visitas'].mean()))

# Eliminar registros con Email vacío y duplicados
df.dropna(subset=['Email'], inplace=True)
df.drop_duplicates(inplace=True)
df.reset_index(drop=True, inplace=True)

# Guardar el archivo limpio
df.to_csv('DatosPersonalesLimpios.csv', index=False)

# Mostrar final
df
```

```python
import pandas as pd
import numpy as np

# Cargar los datos
df = pd.read_csv('Datosucios.csv')
print("Data original:")
print(df)

# Reemplazar valores None y 'nan' por np.nan
df = df.replace([None, 'nan'], np.nan)
print("\nData después de reemplazar None y 'nan':")
print(df)

# Imputar valores faltantes correctamente
df['Edad'] = df['Edad'].fillna(df['Edad'].median())
df['Ingresos'] = df['Ingresos'].fillna(df['Ingresos'].median())
df['Visitas'] = df['Visitas'].fillna(int(df['Visitas'].mean()))

print("\nData después de imputar valores faltantes:")
print(df)

# Eliminar registros sin Email
df = df.dropna(subset=['Email'])
print("\nData después de eliminar registros sin Email:")
print(df)

# Resetear el índice
df = df.reset_index(drop=True)

# Guardar los datos limpios
cleaned_data_path = 'DatosLimpios.csv'
df.to_csv(cleaned_data_path, index=False)

print("\nData final limpia y guardada:")
print(df)

# Mostrar ruta del archivo guardado
print("\nArchivo guardado en:", cleaned_data_path)
```

```python
import pandas as pd
import numpy as np

# Crear el dataset
data = {
    'Nombre': ['Juan', 'Maria', 'Pedro', None, 'Ana', 'Carlos', 'Juan'],
    'Edad': [25, 30, np.nan, 22, 28, np.nan, 25],
    'Genero': ['M', 'F', 'M', None, 'F', 'M', 'M'],
    'Email': ['juan@mail.com', 'maria@mail.com', 'pedro@mail.com', 'ana@mail.com', 'ana@mail.com', None, 'juan@mail.com'],
    'Pais': ['Colombia', None, 'Mexico', 'Argentina', 'Argentina', 'Brasil', 'Colombia']
}

df = pd.DataFrame(data)
print("Data original:")
print(df)

# Mostrar valores nulos
print("Valores faltantes:\n", df.isnull().sum())

df['Edad'] = df['Edad'].fillna(df['Edad'].median())  # Mediana para Edad
df['Genero'] = df['Genero'].fillna('Desconocido')    # Valor específico para Género
df['Pais'] = df['Pais'].fillna(df['Pais'].mode()[0])  # Moda para País

# 2. Estandarizar Género
df['Genero'] = df['Genero'].replace({'M': 'Masculino', 'F': 'Femenino'})

# 3. Eliminar duplicados
df = df.drop_duplicates()

# 4. Eliminar filas sin Email
df = df.dropna(subset=['Email'])

# 5. Resetear el índice
df = df.reset_index(drop=True)

# Mostrar dataset limpio
print("\nData limpia:")
print(df)
```

```python
import pandas as pd

data = {
    'Nombre': ['Carlos', 'Ana', 'Luis', 'Pedro', 'María', None],
    'Edad': [34, 29, None, 45, 25, 33],
    'FechaIngreso': ['2021-01-15', '2020/03/12', '2022-07-01', '2021/12/01', '2021-05-20', '2020-03-12'],
    'ResultadoPrueba': ['Positivo', 'Negativo', 'Positivo', 'Positivo', 'Negativo', 'Negativo']
} 
df = pd.DataFrame(data)
print("Data original:")
print(df)
print("\nValores faltantes:")
print(df.isnull().sum())

df['FechaIngreso'] = pd.to_datetime(df['FechaIngreso'], errors='coerce',format='%Y-%m-%d')
df['Edad'] = df['Edad'].fillna(df['Edad'].median())
df = df.dropna(subset=['Nombre'])
print("\nData después de convertir FechaIngreso:")
print(df)
```

```python
import pandas as pd
data = { 'ClienteID': [1, 2, 3, 4, 5],
        'Producto': ['A', 'B', 'C', 'D', 'E'],
         'Cantidad': [10, 20, None, 5, 15],
         'Precio': ['100', '200$', '300', 'cuatrocientos', '500'],
         'FechaTransaccion': ['2023-01-15', '2023/02/20', '2023-03-25', '2023/04/30', '2023-05-10']
}
df = pd.DataFrame(data)
print("Data original:")
print(df)
df['Cantidad'] = df['Cantidad'].fillna(df['Cantidad'].median())
df['Precio'] = df['Precio'].replace('cuatrocientos', '400')
df['Precio'] = df['Precio'].str.replace(r'[^0-9]', '', regex=True)
df['FechaTransaccion'] = pd.to_datetime(df['FechaTransaccion'], errors='coerce',format='%Y-%m-%d')
print(df)
```

```python
import pandas as pd

# Crear un DataFrame con los datos sucios
data = {
    'Nombre': ['Juan', 'Maria', 'Pedro', 'Ana', 'Luisa', 'Carlos', 'Juan'],
    'Edad': [25, 30, None, 29, 28, 200, 25],
    'Email': ['juan@mail.com', 'maria@mail.com', 'pedro@mail.com', 'ana@mail.com',
              'luisa@mail.com', 'carlos@mail.com', 'juan@mail.com'],
    'Pais': ['colombia', 'Colombia', 'Mexico', 'Argentina', 'Colombia', 'Brazil', 'colombia'],
    'Genero': ['M', 'F', 'H', 'F', 'Femenino', 'M', 'M'],
    'Visitas': [5, 7, 6, 4, 8, 2, 5]
}
df = pd.DataFrame(data)
print("Data original:")
print(df)
df['Edad'] = df['Edad'].fillna(df['Edad'].median())
df['Edad'] = df['Edad'].replace({
    'M': 'Masculino',
    'H': 'Masculino',
    'F': 'Femenino',
    'Femenino': 'Femenino'
})
df['Pais'] = df['Pais'].str.capitalize()
df.loc[df['Edad'] > 100, 'Edad'] = df['Edad'].median()
print("valores despues de la limpieza")
print(df)
```
